```python
import pandas as pd
import os
import shutil
```

```python
# creating data for positive samples

FILE_PATH = "github-chestxray-dataset/metadata.csv"
IMAGES_PATH = "github-chestxray-dataset/images"
```

```python
dataframe = pd.read_csv(FILE_PATH)
dataframe.head()
```

```python
COVID_IMAGES = "Dataset/Covid"
```

```python
count = 0

# iterating through dataframe and picking out PA scans with COVID-19 to add to dataset folder

for (i, row) in dataframe.iterrows():
    if "COVID-19" in row["finding"] and row["view"] == "PA":
        filename = row["filename"]
        image_path = os.path.join(IMAGES_PATH, filename)
        target_path = os.path.join(COVID_IMAGES, filename)
        shutil.copy2(image_path, target_path)
        
        count += 1
        print("image", count, "added to covid images folder")
```

```python
import random
KAGGLE_PATH = "kaggle-chest-xray/train/NORMAL"
```

```python
NORMAL_IMAGES = "Dataset/Normal"
normal_images_name = os.listdir(KAGGLE_PATH)
random.shuffle(normal_images_name)

for i in range(142):
    image_name = normal_images_name[i]
    image_path = os.path.join(KAGGLE_PATH, image_name)
    target_path = os.path.join(NORMAL_IMAGES, image_name)
    shutil.copy2(image_path, target_path)
    print("image", i, "added to normal images folder")
```
